```python
import matplotlib.pyplot as plt
```

```python
#Plot distribution of power
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'], bins=30, kde=True)
plt.title('Distribution of Generated Power (kw)')
plt.xlabel('Generated Power (kw)')
plt.ylabel('Frequency')
plt.show()
```

```python
df[df.columns[18:20]].hist(bins=30, figsize=(15, 5))
plt.show()
```

```python
# Bivariate analysis
# scatter plot with target feature
plt.figure(figsize=(15, 30))
for i, colum in enumerate(df.columns):
    if(i != 20):
        plt.subplot(7, 3, i+1)
        plt.scatter(df[colum], df['generated_power_kw'])
        plt.title(f'{colum} vs Generated Power (kw)')
        plt.xlabel(colum)
        plt.ylabel('Generated Power (kw)')
plt.tight_layout()
plt.show()
```

```python
# Co-relation Matrix
df.corr()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=True, fmt='.2f')
```

```python
# Outlier
plt.figure(figsize=(15, 30))
for i, colum in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    plt.boxplot(df[colum])
plt.show()
```
